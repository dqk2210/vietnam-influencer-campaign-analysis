## Step 5. Clean Column Names

```python
def clean_column_name(col):
    col = col.strip().lower()
    col = re.sub(r"[^a-z0-9]+", "_", col)
    col = re.sub(r"_+", "_", col).strip("_")
    return col

df = df_raw.copy()
df.columns = [clean_column_name(c) for c in df.columns]

df.head()
```

## Step 6. Convert Text to Number

```python
def parse_compact_number(value):
    if pd.isna(value):
        return np.nan
    
    s = str(value).strip().upper()
    s = s.replace("\xa0", "").replace(" ", "")
    
    multiplier = 1
    
    if s.endswith("K"):
        multiplier = 1_000
        s = s[:-1]
    elif s.endswith("M"):
        multiplier = 1_000_000
        s = s[:-1]
    elif s.endswith("B"):
        multiplier = 1_000_000_000
        s = s[:-1]
    
    s = s.replace(",", ".")
    s = re.sub(r"[^0-9.\-]", "", s)
    
    try:
        return float(s) * multiplier
    except:
        return np.nan


def parse_percent(value):
    if pd.isna(value):
        return np.nan
    
    s = str(value).replace("%", "").replace(",", ".")
    s = re.sub(r"[^0-9.\-]", "", s)
    
    try:
        return float(s)
    except:
        return np.nan


def parse_price(value):
    if pd.isna(value):
        return np.nan
    
    s = str(value).upper().replace("VND", "")
    s = re.sub(r"[^0-9]", "", s)
    
    try:
        return float(s)
    except:
        return np.nan
```

```python
df["followers_num"] = df["followers"].apply(parse_compact_number)
df["median_views_num"] = df["median_views"].apply(parse_compact_number)
df["engagement_pct"] = df["engagement"].apply(parse_percent)
df["price_num"] = df["price"].apply(parse_price)

for col in ["collab_score", "broadcast_score", "commerce_score"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

## Step 7. Handle Duplicate Influencers

```python
def mode_or_first(series):
    s = series.dropna().astype(str)
    if s.empty:
        return np.nan
    return s.mode().iloc[0] if len(s.mode()) > 0 else s.iloc[0]


def combine_unique_text(series):
    values = []
    for item in series.dropna():
        parts = [x.strip() for x in str(item).split(",") if x.strip()]
        values.extend(parts)
    
    values = sorted(set(values))
    return ", ".join(values) if values else np.nan
```

```python
agg_rules = {
    "name": mode_or_first,
    "country": mode_or_first,
    "collab_score": "median",
    "broadcast_score": "median",
    "commerce_score": "median",
    "followers_num": "median",
    "median_views_num": "median",
    "engagement_pct": "median",
    "price_num": "median",
    "field_tags": combine_unique_text,
    "video_tags": combine_unique_text
}

df_cleaned = (
    df.dropna(subset=["id"])
    .groupby("id", as_index=False)
    .agg(agg_rules)
)

print("Before:", df.shape)
print("After:", df_cleaned.shape)
```

## Step 8. Feature Engineering

```python
df_cleaned["cpv"] = (
    df_cleaned["price_num"] / df_cleaned["median_views_num"].replace(0, np.nan)
)

df_cleaned["view_rate_pct"] = (
    df_cleaned["median_views_num"] / df_cleaned["followers_num"].replace(0, np.nan) * 100
)
```

```python
conditions = [
    df_cleaned["followers_num"] < 10_000,
    df_cleaned["followers_num"].between(10_000, 99_999),
    df_cleaned["followers_num"].between(100_000, 499_999),
    df_cleaned["followers_num"].between(500_000, 999_999),
    df_cleaned["followers_num"] >= 1_000_000
]

choices = ["Nano", "Micro", "Mid-tier", "Macro", "Mega"]

df_cleaned["creator_segment"] = np.select(
    conditions,
    choices,
    default="Unknown"
)
```

#### Price tier:

```python
price_conditions = [
    df_cleaned["price_num"].isna(),
    df_cleaned["price_num"] < 5_000_000,
    df_cleaned["price_num"].between(5_000_000, 14_999_999),
    df_cleaned["price_num"].between(15_000_000, 49_999_999),
    df_cleaned["price_num"] >= 50_000_000
]

price_choices = [
    "Unknown",
    "Low",
    "Medium",
    "High",
    "Premium"
]

df_cleaned["price_tier"] = np.select(
    price_conditions,
    price_choices,
    default="Unknown"
)
```

## Step 9. Create Potential Group

```python
thresholds = {
    "broadcast_high": df_cleaned["broadcast_score"].quantile(0.75),
    "commerce_high": df_cleaned["commerce_score"].quantile(0.75),
    "collab_high": df_cleaned["collab_score"].quantile(0.75),
    "engagement_high": df_cleaned["engagement_pct"].quantile(0.75),
    "engagement_mid": df_cleaned["engagement_pct"].quantile(0.50),
    "views_high": df_cleaned["median_views_num"].quantile(0.75),
    "views_mid": df_cleaned["median_views_num"].quantile(0.50),
    "cpv_good": df_cleaned["cpv"].quantile(0.25),
    "cpv_mid": df_cleaned["cpv"].quantile(0.50),
    "cpv_high": df_cleaned["cpv"].quantile(0.75)
}
```

```python
def classify_potential(row):
    has_commerce = pd.notna(row["commerce_score"])
    
    if (
        row["broadcast_score"] >= thresholds["broadcast_high"]
        and has_commerce
        and row["commerce_score"] >= thresholds["commerce_high"]
        and row["collab_score"] >= thresholds["collab_high"]
    ):
        return "All-round Potential"
    
    elif (
        has_commerce
        and row["commerce_score"] >= thresholds["commerce_high"]
        and row["engagement_pct"] >= thresholds["engagement_high"]
        and row["cpv"] <= thresholds["cpv_mid"]
    ):
        return "Sales Potential"
    
    elif (
        row["broadcast_score"] >= thresholds["broadcast_high"]
        and row["median_views_num"] >= thresholds["views_high"]
    ):
        return "Awareness Booster"
    
    elif (
        row["collab_score"] >= thresholds["collab_high"]
        and row["engagement_pct"] >= thresholds["engagement_mid"]
    ):
        return "Collab-ready"
    
    elif (
        row["cpv"] <= thresholds["cpv_good"]
        and row["median_views_num"] >= thresholds["views_mid"]
    ):
        return "Cost-efficient"
    
    elif (
        row["broadcast_score"] >= thresholds["broadcast_high"]
        and row["cpv"] >= thresholds["cpv_high"]
    ):
        return "High Reach - Low Efficiency"
    
    else:
        return "General / Low Priority"


df_cleaned["potential_group"] = df_cleaned.apply(classify_potential, axis=1)
```
